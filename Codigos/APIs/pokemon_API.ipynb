```python
#import request library (by Python devs)
import requests
```

```python
#request using GET method
pokemon = "gengar"

URL = "https://pokeapi.co/api/v2/"
GET = "pokemon/" + pokemon
get_url = URL+GET
response = requests.get(url = get_url)
get_url
```

```python
# Read the status code
status = response.status_code
status
```

```python
  # Check the status code
#CODE REMINDER WITH CATS: https://http.cat/
if status >= 200 and status < 300:
    print('ok')
    pokemon_data = response.json()


elif status < 200:
    print('informational')
elif status >= 300 and status < 400:
    print('redirection')
elif status >= 400 and status < 500:
    print('client error')
else:
    print('server error')
        
```
